````python

agente_busca = Agent(
    name="AgenteBusca",
    model=MODEL_ID,
    instruction="""
    Você é um agente de busca especializado em encontrar oportunidades de emprego, cursos e programas de capacitação.
    Sua principal função é analisar a solicitação do usuário e identificar os termos de busca relevantes.

    **Suas principais responsabilidades:**

    1.  **Análise da Solicitação:**
        * Receba a mensagem do usuário e identifique os termos-chave que indicam sua necessidade (por exemplo, "vagas de programador", "curso de inglês", "bolsa de estudos").

    2.  **Extração de Termos de Busca:**
        * Extraia apenas os termos mais importantes para a busca, omitindo palavras irrelevantes ou de ligação.
        * Retorne os termos separados por vírgula.

    **Exemplo de Entrada:**
    Usuário: "Quero encontrar vagas de emprego para desenvolvedor Python em São Paulo, e também quero saber se tem algum curso online de Machine Learning."

    **Exemplo de Saída:**
    desenvolvedor Python, vagas, São Paulo, curso online, Machine Learning
    """,
    description="Agente responsável por extrair termos de busca de oportunidades de emprego e capacitação.",
    tools=[],
)

class EmpregaBotCoordinator:
    def __init__(self):
        self.onboarding = AgenteOnboardingWrapper()
        self.perfil_completo = False

    async def responder(self, mensagem: str, user_id: str, session_id: str):
        if not self.perfil_completo:
            resposta = self.onboarding.processar_mensagem(mensagem, user_id, session_id)
            if "seu perfil foi salvo com sucesso" in resposta.lower():
                self.perfil_completo = True
            return resposta

        busca_input = agente_busca.to_input(mensagem)
        busca_resposta = await agente_busca.run_async(input=busca_input)
        termos = busca_resposta.text.strip()

        return f"Entendi! Estou buscando oportunidades relacionadas a: {termos}"

    def resetar_onboarding(self):
        self.onboarding.reset()
        self.perfil_completo = False


````

````python
# ========================= Wrappers dos Agentes =========================

def call_agent(agent: Agent, user_id: str, session_id: str, message_text: str) -> str:
    """Envia uma mensagem para o agente via ADK Runner e retorna a resposta completa."""
    try:
        # conn = sqlite3.connect('/content/drive/MyDrive/empregabot.py.db')
        # cursor = conn.cursor()
        # cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
        # print(cursor.fetchall())  # Deveria mostrar [('curso',), ('vaga',)]

        # session_service.create_session(app_name=agent.name, user_id=user_id, session_id=session_id)
        content = types.Content(role="user", parts=[types.Part(text=message_text)])
        runner = Runner(app_name=agent.name, agent=agent, session_service=session_service)
        final_response = []
        for event in runner.run(user_id=user_id, session_id=session_id, new_message=content):
            if event.is_final_response():
                for part in event.content.parts:
                    if part.text:
                        final_response.append(part.text)
        return "\n".join(final_response)
    except Exception as e:
        print(f"[Error] Agent {agent.name} failed: {e}")
        return "Desculpe, ocorreu um erro ao processar sua solicitação. Por favor, tente novamente mais tarde."


class AgenteOnboardingWrapper:
    def __init__(self):
        # Lista de campos que queremos coletar, na ordem
        self.campos = [
            ("nome", "Qual é o seu nome completo?"),
            ("idade", "Quantos anos você tem?"),
            ("email", "Qual é o seu e-mail para contato?"),
            ("escolaridade", "Qual o seu nível de escolaridade?"),
            ("habilidades", "Quais são suas principais habilidades (técnicas e comportamentais)?"),
            ("interesses", "Em quais áreas você tem mais interesse de atuar?"),
            ("localizacao", "Onde você mora (Cidade, Estado)?")
        ]
        self.perfil = {}
        self.etapa_atual = 0

    def processar_mensagem(self, mensagem: str, user_id: str, session_id: str) -> str:
        # Se for a primeira interação ou depois da saudação:
        if self.etapa_atual == 0 and mensagem.lower().startswith("olá"):
            prompt = "Olá! " \
                     "Antes de eu recomendar oportunidades, preciso fazer algumas perguntas."
            self.etapa_atual = 1
            return prompt + "\n\n" + self.campos[0][1]

        # Para as próximas etapas:
        if 1 <= self.etapa_atual <= len(self.campos):
            campo, pergunta = self.campos[self.etapa_atual - 1]
            # Armazena a resposta bruta
            self.perfil[campo] = mensagem.strip()
            # Avança
            if self.etapa_atual < len(self.campos):
                self.etapa_atual += 1
                # Pergunta o próximo campo
                return self.campos[self.etapa_atual - 1][1]
            else:
                # Todas respostas coletadas → finaliza
                self.etapa_atual += 1
                return self._finalizar_onboarding()

        # Qualquer outra entrada fora de fluxo:
        return "Vamos recomeçar? Digite 'Olá' para iniciar seu cadastro."

    def _finalizar_onboarding(self) -> str:
        # Aqui você salva self.perfil no seu serviço de persistência
        # Exemplo: usuario_service.salvar_perfil(self.perfil)
        return (
            f"Obrigado(a), {self.perfil.get('nome', 'jovem')}! "
            "Seu perfil foi salvo com sucesso. Agora posso te ajudar a encontrar oportunidades. "
            "O que gostaria de fazer agora?"
        )

    def reset(self):
        self.perfil = {}
        self.etapa_atual = 0


class AgenteBuscaWrapper:
    def __init__(self):
        self.busca_service = BuscaService()  # Inicializa o serviço de busca

    def processar_mensagem(self, mensagem: str, user_id: str, session_id: str) -> str:
        """Processa a solicitação de busca do usuário."""
        try:
            # O agente de busca deve ser capaz de interpretar a mensagem e extrair os termos de busca
            prompt = f"Com base na seguinte mensagem do usuário, identifique os termos de busca relevantes para oportunidades de emprego ou capacitação. Retorne apenas os termos separados por vírgula. Mensagem: {mensagem}"
            termos_busca_str = call_agent(agente_busca, user_id, session_id, prompt)
            termos_busca = [termo.strip() for termo in termos_busca_str.split(',') if termo.strip()]

            if not termos_busca:
                return "Não consegui identificar o que você procura. Poderia me dizer de outra forma?"

            # Realiza a busca usando os termos extraídos (adaptado para usar o BuscaService)
            perfil_temporario = {"interesses": termos_busca, "localizacao": ""}  # Cria um perfil temporário para a busca
            resultados = self.busca_service.buscar_oportunidades(perfil_temporario, user_id, session_id)

            if "Não encontrei oportunidades relevantes" in resultados:
                return f"Não encontrei muitas oportunidades para '{', '.join(termos_busca)}'. Que tal tentar outros termos?"

            return f"Encontrei as seguintes oportunidades para '{', '.join(termos_busca)}':\n\n{resultados}"

        except Exception as e:
            print(f"[Error] Agent {agente_busca.name} failed during search: {e}")
            return "Desculpe, ocorreu um erro ao buscar oportunidades. Por favor, tente novamente mais tarde."


class AgenteConversacionalWrapper:
    def processar_mensagem(self, mensagem: str, user_id: str, session_id: str) -> str:
        """Envia a mensagem diretamente para o agente conversacional."""
        try:
            response = call_agent(agente_conversacional, user_id, session_id, mensagem)
            return response
        except Exception as e:
            print(f"[Error] Agent {agente_conversacional.name} failed: {e}")
            return "Desculpe, tive um problema para te responder. Pode repetir, por favor?"


class AgenteRecomendacaoWrapper:
    def __init__(self):
        self.recomendacao_service = RecomendacaoService()  # Inicializa o serviço de recomendação

    def recomendar_oportunidades(self, perfil, user_id: str, session_id: str) -> str:
        """Gera recomendações de vagas, cursos e programas com base no perfil do usuário."""
        try:
            recomendacoes = self.recomendacao_service.gerar_recomendacoes(perfil, user_id, session_id)
            return f"Com base no seu perfil, recomendo as seguintes oportunidades:\n\n{recomendacoes}"
        except Exception as e:
            print(f"[Error] Agent {agente_recomendacao.name} failed during recommendation: {e}")
            return "Desculpe, ocorreu um erro ao gerar recomendações. Por favor, tente novamente mais tarde."


class AgenteEntrevistaWrapper:
    def simular_entrevista(self, perfil, user_id: str, session_id: str) -> str:
        """Simula uma entrevista de emprego com o usuário."""
        try:
            prompt = f"Simule uma entrevista de emprego para um candidato com o seguinte perfil: {perfil}"
            resposta = call_agent(agente_entrevista, user_id, session_id, prompt)
            return resposta
        except Exception as e:
            print(f"[Error] Agent {agente_entrevista.name} failed during interview simulation: {e}")
            return "Desculpe, ocorreu um erro ao simular a entrevista. Por favor, tente novamente mais tarde."

class AgenteCurriculoWrapper:
    def revisar_curriculo(self, curriculo, user_id: str, session_id: str) -> str:
        """Revisa o currículo do usuário e fornece feedback."""
        try:
            prompt = f"Analise o seguinte currículo e forneça sugestões de melhoria: {curriculo}"
            resposta = call_agent(agente_curriculo, user_id, session_id, prompt)
            return resposta
        except Exception as e:
            print(f"[Error] Agent {agente_curriculo.name} failed during resume review: {e}")
            return "Desculpe, ocorreu um erro ao revisar o currículo. Por favor, tente novamente mais tarde."
````

````python
# ========================= Serviços =========================

class BuscaService:
    def __init__(self):
        self.conn = sqlite3.connect('/content/drive/MyDrive/empregabot.py.db')
        self.cursor = self.conn.cursor()

    def buscar_oportunidades(self, perfil, user_id: str, session_id: str) -> str:
        """Busca vagas, cursos e programas de capacitação com base nos termos de busca."""
        try:
            termos_busca = perfil.get("interesses", [])
            resultados = []

            if termos_busca:
                termos_busca_str = ', '.join(f"'{termo}'" for termo in termos_busca)

                # Buscar vagas
                self.cursor.execute(f"""
                    SELECT titulo, descricao, localizacao, salario FROM vaga
                    WHERE INSTR(LOWER(titulo), LOWER({termos_busca_str}))
                       OR INSTR(LOWER(descricao), LOWER({termos_busca_str}))
                """)
                vagas = self.cursor.fetchall()
                if vagas:
                    resultados.append("Vagas:")
                    for vaga in vagas:
                        resultados.append(f"- {vaga[0]} ({vaga[2]}) - Salário: {vaga[3]}")
                else:
                    resultados.append("Não encontrei vagas relevantes.")

                # Buscar cursos
                self.cursor.execute(f"""
                    SELECT titulo, descricao, instituicao FROM curso
                    WHERE INSTR(LOWER(titulo), LOWER({termos_busca_str}))
                       OR INSTR(LOWER(descricao), LOWER({termos_busca_str}))
                """)
                cursos = self.cursor.fetchall()
                if cursos:
                    resultados.append("\nCursos:")
                    for curso in cursos:
                        resultados.append(f"- {curso[0]} ({curso[1]}) - {curso[2]}")
                else:
                    resultados.append("\nNão encontrei cursos relevantes.")
            else:
                resultados.append("Por favor, me diga o que você está procurando.")

            return "\n".join(resultados)

        except Exception as e:
            print(f"[Error] Error during opportunity search: {e}")
            return "Desculpe, ocorreu um erro ao buscar oportunidades."

        finally:
            self.conn.close()


class RecomendacaoService:
    def __init__(self):
        self.conn = sqlite3.connect('/content/drive/MyDrive/empregabot.py.db')
        self.cursor = self.conn.cursor()

    def gerar_recomendacoes(self, perfil, user_id: str, session_id: str) -> str:
        """Gera recomendações personalizadas para o usuário."""
        try:
            termos_interesse = perfil.get("interesses", [])
            localizacao = perfil.get("localizacao", "")
            resultados = []

            if termos_interesse:
                termos_interesse_str = ', '.join(f"'{termo}'" for termo in termos_interesse)

                # Recomendar vagas
                self.cursor.execute(f"""
                    SELECT titulo, descricao, localizacao, salario FROM vaga
                    WHERE INSTR(LOWER(titulo), LOWER({termos_interesse_str}))
                       OR INSTR(LOWER(descricao), LOWER({termos_interesse_str}))
                """)
                vagas = self.cursor.fetchall()
                if vagas:
                    resultados.append("Vagas Recomendadas:")
                    for vaga in vagas:
                        resultados.append(f"- {vaga[0]} ({vaga[2]}) - Salário: {vaga[3]}")
                else:
                    resultados.append("Não encontrei vagas que correspondam aos seus interesses.")

                # Recomendar cursos
                self.cursor.execute(f"""
                    SELECT titulo, descricao, instituicao FROM curso
                    WHERE INSTR(LOWER(titulo), LOWER({termos_interesse_str}))
                       OR INSTR(LOWER(descricao), LOWER({termos_interesse_str}))
                """)
                cursos = self.cursor.fetchall()
                if cursos:
                    resultados.append("\nCursos Recomendados:")
                    for curso in cursos:
                        resultados.append(f"- {curso[0]} ({curso[1]}) - {curso[2]}")
                else:
                    resultados.append("\nNão encontrei cursos que correspondam aos seus interesses.")

            else:
                resultados.append("Por favor, me diga quais são seus interesses para que eu possa gerar recomendações.")

            return "\n".join(resultados)

        except Exception as e:
            print(f"[Error] Error generating recommendations: {e}")
            return "Desculpe, ocorreu um erro ao gerar recomendações."

        finally:
            self.conn.close()
````

````python
# ========================= EmpregaBot =========================

class EmpregaBot:
    def __init__(self):
        self.agente_onboarding_wrapper = AgenteOnboardingWrapper()
        self.agente_busca_wrapper = AgenteBuscaWrapper()
        self.agente_conversacional_wrapper = AgenteConversacionalWrapper()
        self.agente_recomendacao_wrapper = AgenteRecomendacaoWrapper()
        self.agente_entrevista_wrapper = AgenteEntrevistaWrapper()
        self.agente_curriculo_wrapper = AgenteCurriculoWrapper()

        self.user_id = "user123"  # Substitua por um ID de usuário real
        self.session_id_onboarding = "onboarding_session"
        self.session_id_busca = "busca_session"
        self.session_id_conversacional = "conversacional_session"
        self.session_id_recomendacao = "recomendacao_session"
        self.session_id_entrevista = "entrevista_session"
        self.session_id_curriculo = "curriculo_session"

        self.perfil_usuario = None
        self.estado_conversa = "onboarding"  # Estado inicial da conversa

    def iniciar(self):
        print("Bem-vindo ao EmpregaBot! Como posso te ajudar hoje?")
        while True:
            mensagem = input("Você: ")
            if mensagem.lower() == "sair":
                print("EmpregaBot: Até logo!")
                break

            resposta = self.processar_mensagem(mensagem)
            print(f"\nEmpregaBot:\n{resposta}")

    def processar_mensagem(self, mensagem: str) -> str:
        """Processa a mensagem do usuário e direciona para o agente apropriado."""

        try:
            if self.estado_conversa == "onboarding":
                resposta = self.agente_onboarding_wrapper.processar_mensagem(
                    mensagem, self.user_id, self.session_id_onboarding
                )
                if "Obrigado(a)" in resposta:  # Verificar se o onboarding foi concluído
                    self.estado_conversa = "aguardando_acao"
                    self.perfil_usuario = self.agente_onboarding_wrapper.perfil
                    print("\nPerfil do usuário coletado:", self.perfil_usuario)  # Log do perfil
                return resposta

            elif self.estado_conversa == "aguardando_acao":
                if "vagas" in mensagem.lower() or "emprego" in mensagem.lower() or "buscar" in mensagem.lower():
                    return self.agente_busca_wrapper.processar_mensagem(
                        mensagem, self.user_id, self.session_id_busca
                    )
                elif "cursos" in mensagem.lower() or "capacitação" in mensagem.lower():
                    return self.agente_busca_wrapper.processar_mensagem(
                        mensagem, self.user_id, self.session_id_busca  # Usando BuscaWrapper para simplificar
                    )
                elif "recomendações" in mensagem.lower() or "oportunidades" in mensagem.lower():
                    if self.perfil_usuario:
                        return self.agente_recomendacao_wrapper.recomendar_oportunidades(
                            self.perfil_usuario, self.user_id, self.session_id_recomendacao
                        )
                    else:
                        return "Por favor, complete seu perfil primeiro para receber recomendações."
                elif "entrevista" in mensagem.lower():
                    if self.perfil_usuario:
                        return self.agente_conversacional_wrapper.processar_mensagem(
                            mensagem, self.user_id, self.session_id_conversacional
                        )  # Usando Conversacional para simplificar
                    else:
                        return "Por favor, complete seu perfil primeiro para simular uma entrevista."
                elif "currículo" in mensagem.lower() or "curriculo" in mensagem.lower():
                    return self.agente_conversacional_wrapper.processar_mensagem(
                        mensagem, self.user_id, self.session_id_conversacional
                    )  # Usando Conversacional para simplificar
                elif "olá" in mensagem.lower() or "oi" in mensagem.lower() or "ola" in mensagem.lower():
                    return self.agente_conversacional_wrapper.processar_mensagem(
                        mensagem, self.user_id, self.session_id_conversacional
                    )
                else:
                    return self.agente_conversacional_wrapper.processar_mensagem(
                        mensagem, self.user_id, self.session_id_conversacional
                    )

            else:
                return "Desculpe, não entendi. Por favor, tente novamente."

        except Exception as e:
            print(f"[Error] Error processing message: {e}")
            return "Desculpe, ocorreu um erro ao processar sua mensagem."

    def _executar_revisao_curriculo(self):
        if self.perfil_usuario:
            resposta = self.agente_curriculo_wrapper.revisar_curriculo(
                self.perfil_usuario, self.user_id, self.session_id_curriculo
            )
            print(f"\nEmpregaBot:\n{resposta}")
        else:
            print("Por favor, complete seu perfil primeiro.")

    def _executar_simulacao_entrevista(self):
        if self.perfil_usuario:
            resposta = self.agente_entrevista_wrapper.simular_entrevista(
                self.perfil_usuario, self.user_id, self.session_id_entrevista
            )
            print(f"\nEmpregaBot: {resposta}")
        else:
            print("Por favor, complete seu perfil primeiro.")


# ========================= Execução do EmpregaBot =========================

if __name__ == "__main__":
    bot = EmpregaBot()
    bot.iniciar()
````
